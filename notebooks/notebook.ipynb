```python
from pathlib import Path

from IPython.display import Image
from IPython.display import display as ipy_display

for img_path in sorted(Path().glob("benchmark_*.png")):
    ipy_display(img_path.name)
    ipy_display(Image(filename=str(img_path)))
```

```python
import base64
from pathlib import Path

from IPython.display import HTML
from IPython.display import display as ipy_display

for img_path in sorted(Path().glob("benchmark_*.png")):
    with img_path.open("rb") as file_obj:
        b64 = base64.b64encode(file_obj.read()).decode()
    ipy_display(
        HTML(
            f'<a href="data:image/png;base64,{b64}" download="{img_path.name}">'
            f"Download {img_path.name}</a>"
        )
    )
```
